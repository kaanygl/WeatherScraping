```python
import pandas as pd
import numpy as np
```

```python
df_brown = pd.read_csv("/content/Indoor Air Quality Monitoring @ Brindisi Airport/complete_measure_node_airport_brown.csv",header=None)
df_brown = df_brown.drop(index=0).reset_index(drop=True)
```

```python
df_brown
```

```python
split_data_brown = df_brown.iloc[:, 0].apply(lambda x: x.split(";"))
```

```python

columns_brown = "id_measure;node_id;ads1115_value;ads1115_voltage;scd30_co2;scd30_temp;scd30_hum;sps30_pm05_count;sps30_pm1_count;sps30_pm1_ug;sps30_pm25_count;sps30_pm25_ug;sps30_pm4_count;sps30_pm4_ug;sps30_pm10_count;sps30_pm10_ug;sps30_pm_typ;ts_insertion"

column_names_brown = columns_brown.split(";")

print(column_names_brown)
```

```python
split_data_brown
```

```python
reformatted_df_brown = pd.DataFrame(split_data_brown.tolist(), columns=column_names_brown)
```

```python
reformatted_df_brown
```

```python
column_dtypes = {
    'id_measure': int,
    'node_id': str,
    'ads1115_value': int,
    'ads1115_voltage': float,
    'scd30_co2': float,
    'scd30_temp': float,
    'scd30_hum': float,
    'sps30_pm05_count': float,
    'sps30_pm1_count': float,
    'sps30_pm1_ug': float,
    'sps30_pm25_count': float,
    'sps30_pm25_ug': float,
    'sps30_pm4_count': float,
    'sps30_pm4_ug': float,
    'sps30_pm10_count': float,
    'sps30_pm10_ug': float,
    'sps30_pm_typ': float,
    'ts_insertion': str
}
```

```python
for i in reformatted_df_brown.columns:
  if type(i) == float:
    reformatted_df_brown[i] = reformatted_df_brown[i].replace('NULL', np.nan)
    reformatted_df_brown[i] = reformatted_df_brown[i].astype(float)
  elif type(i) == int:
    reformatted_df_brown[i] = reformatted_df_brown[i].replace('NULL', np.nan)
    reformatted_df_brown[i] = reformatted_df_brown[i].astype(int)
  else:
    reformatted_df_brown[i] = reformatted_df_brown[i].replace('NULL', np.nan)
    reformatted_df_brown[i] = reformatted_df_brown[i].astype(str)
```

```python
reformatted_df_brown['id_measure'] = pd.to_numeric(reformatted_df_brown['id_measure'], errors='coerce')
reformatted_df_brown['id_measure'] = reformatted_df_brown['id_measure'].fillna(0).astype(int)
```

```python
reformatted_df_brown = reformatted_df_brown.astype(column_dtypes)
```

```python
output_file_brown = "processed_brown.csv"
reformatted_df_brown.to_csv(output_file_brown, index=False)
```
